```python
import requests
 
def download_url(meetDate):
    if meetDate.month < 10 and meetDate.day < 10:
        meetDate = str(meetDate.year)+'_0'+str(meetDate.month)+'_0'+str(meetDate.day)
    elif meetDate.month < 10:
        meetDate = str(meetDate.year)+'_0'+str(meetDate.month)+'_'+str(meetDate.day)
    elif meetDate.day < 10:
        meetDate = str(meetDate.year)+'_'+str(meetDate.month)+'_0'+str(meetDate.day)
    else:
        meetDate = str(meetDate.year)+'_'+str(meetDate.month)+'_'+str(meetDate.day)
    r = requests.get('https://www.noisebridge.net/wiki/Meeting_Notes_'+ meetDate)# Making a GET request

# print content of request
    with open(r'C:\Users\remya\OneDrive\Documents\MSDA Spring 2023\NewMN_dwnlds\Meeting_Notes_'+meetDate + '.html', 'wb') as html_file:
        html_file.write(r.content)
```

```python
# import datetime module
import datetime
 
# consider the start date as 2007-December 4 th
start_date = datetime.date(2023, 1, 3) # files has been scraped for 12 months every year
 
# consider the end date as 2023-April 4th
end_date = datetime.date(2023, 4, 4) # latest file that has been scraped as of now. Total 613 HTML files for NB Meeting Notes
 
# delta time
delta = datetime.timedelta(days=7)
 
# iterate over range of dates
while (start_date <= end_date):
    download_url(start_date)
    start_date += delta
```
