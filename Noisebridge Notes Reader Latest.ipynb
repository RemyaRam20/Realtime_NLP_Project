In the following code block we are iterating through all the meeting notes one by one, saving the document name and extracting the discussion text. The html header for the discussion text varies quite a bit from document to document, so we have included a series of if statements that seek to capture the discussion text regardless of the header type. There are however, a number of documents for which the meeting notes have not been captured. A preliminary QA shows that most of these notes have very little information. 

```python
document_name = [] 
text_list = []

for file in os.listdir(directory):
    with open(file, 'r') as fileHandle:
      mht_doc = fileHandle.read()
      soup = BeautifulSoup(mht_doc, 'html.parser')
      document_name.append((os.path.splitext(file)[0]))
      text = []
      for h1 in soup.find_all('h1'):
        if 'Discussion Items' in h1.text or 'Discussion Notes' in h1.text or 'Discussion' in h1.text or 'Excellence' in h1.text:
          for p in h1.find_all_next('p'):
              text.append(p.text)
          for li in h1.find_all_next('li'):
            text.append(li.text)
          for ul in h1.find_all_next('ul'):
            text.append(ul.text)
        elif '2014' in h1.text or '2015' in h1.text or'2016' in h1.text or'2017' in h1.text or'2018' in h1.text or'2019' in h1.text or'2020' in h1.text:
          for p in h1.find_all_next('p'):
            text.append(p.text)
      for h2 in soup.find_all('h2'):
        if 'Meeting Notes' in h2.text or 'Discussion' in h2.text or 'Notes' in h2.text or 'Excellence' in h2.text:
          for p in h2.find_all_next('p'):
              text.append(p.text)
          for li in h2.find_all_next('li'):
            text.append(li.text)
        elif 'Discussion Items' in h2.text:
          for p in h2.find_all_next('p'):
              text.append(p.text)
          for li in h2.find_all_next('li'):
            text.append(li.text)
      for h3 in soup.find_all('h3'):
        if 'Discussion Items' in h3.text:
          for p in h3.find_all_next('p'):
              text.append(p.text)
          for li in h3.find_all_next('li'):
            text.append(li.text)
        for b in soup.find_all('b'):
          if b.text == 'Agenda':
            for p in b.find_all_next('p'):
              text.append(p.text)
      text_list.append(text)
counter = 0
for i in range(len(text_list)):
  if len(text_list[i]) == 0:
    print(document_name[i])
    counter += 1

print(counter)
```

In the following code block we are breaking apart each meeting note's list of text and combining it into one large block of text. This way each meeting note has its own single block of text.

```python
dictionaryOfNotes = {}
for doc, text in zip(document_name, text_list):
    text_clean = []
    for line in text:
      line = line.split(' ')
      tempList = []
      for word in line:
        if '=' not in word:
          tempList.append(word)
      line = ' '.join(tempList)
      text_clean.append(line)
    dictionaryOfNotes[doc] = " ".join(text_clean)
```

```python
 #print(te.get_emotion(' '.join(text_list[30])))

# te.get_emotion("wow this package is really amazing")
sentiment_dictionary = {}
for meeting, text in dictionaryOfNotes.items():
  sentiment_dictionary[meeting] = te.get_emotion(text)
```

```python
sentiment_dictionary2 = {}
for key, val in sentiment_dictionary.items():
  summ = 0
  for score in val.values():
    summ += score
  if summ != 0:
    sentiment_dictionary2[key] = val

len(sentiment_dictionary)
```
